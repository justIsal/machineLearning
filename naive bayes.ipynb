```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

menentukan label dan atribute serta membagi dataset menjadi dua subset yaitu data training dan data testing

```python
x = data.iloc[:,:5]
y = data['play']

onehot_encoder = OneHotEncoder(sparse=False)
label_encoder = LabelEncoder()

label_encoder.fit(y)
x_Train_encoded = onehot_encoder.fit_transform(x)
y_Train_encoded = label_encoder.transform(y)

x_Train,x_Test,y_Train,y_Test = train_test_split(x_Train_encoded,y_Train_encoded,test_size=0.1, random_state=42)
```

```python
# print('dibawah merupakan data x: ')
# print(x)
# print('dibawah merupakan data x_Train :')
```

melatih model menggunakan naive bayes

```python
model = GaussianNB()
model.fit(x_Train,y_Train)
```

```python
x_Test_original = onehot_encoder.inverse_transform(x_Test)
y_Test_original = label_encoder.inverse_transform(y_Test)

print("Data Testing:")
for i, data in enumerate(zip(x_Test_original, y_Test_original)):
    x_data, y_data = data
    print(f"Data ke-{i+1}:")
    print("Features:", x_data)
    print("Target:", y_data)
    print()
```

evaluasi

```python
y_pred = model.predict(x_Test)
accuracy = metrics.accuracy_score(y_Test, y_pred)
predictions = label_encoder.inverse_transform(y_pred)
accuracy_percent = accuracy * 100
print("akurasi: ", accuracy_percent, "%")
print()
print("Hasil Prediksi:")
for i, pred in enumerate(predictions):
    print(f"Data ke-{i+1}: {pred}")
```
